```python
# twitter class

import requests

class Tweet(object):
    def __init__(self, s, embed_str=False):
        if not embed_str:
            # Use Twitter's oEmbed API
            # https://dev.twitter.com/web/embedded-tweets
            api = 'https://publish.twitter.com/oembed?url={}'.format(s)
            response = requests.get(api)
            self.text = response.json()["html"]
        else:
            self.text = s

    def _repr_html_(self):
        return self.text
```

In dieser Illustation ist zu sehen, wie eine Sinuskurve und eine Cosinuskurve zusammen abgebildet sind.

```python
# plot a sine and a cosine

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
z = np.cos(x)

plt.plot(x, y, label="sin(x)")
plt.plot(x, z, label="cos(x)")
plt.legend()

# plot a green cos line with stroke width 10

plt.plot(x, z, color="green", linewidth=10)

# plot a red dashed line

plt.plot(x, z, color="red", linestyle="--")

plt.show()
```
